```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
def new_value(sigma_x, sigma_y, tau_xy, theta, rou):
    plus = (sigma_x+sigma_y)/2
    minus = (sigma_x-sigma_y)/2
    sigma_x1 = round(plus + minus*math.cos(2*math.radians(theta))+tau_xy*math.sin(2*math.radians(theta)),rou)
    print("sigma_x1 =", sigma_x1)
    tau_x1y1 = round(-minus*math.sin(2*math.radians(theta))+tau_xy*math.cos(2*math.radians(theta)),rou)
    print("tau_x1y1 =", tau_x1y1)
    sigma_y1 = round(plus - minus*math.cos(2*math.radians(theta))-tau_xy*math.sin(2*math.radians(theta)),rou)
    print("sigma_y1 =", sigma_y1)
    return (plus, minus, sigma_x1, tau_x1y1, sigma_y1)
```

```python
def rect_plot(theta):
    x0, y0 = -1, -1
    x1 = x0*math.cos(math.radians(theta))-y0*math.sin(math.radians(theta))
    y1 = x0*math.sin(math.radians(theta))+y0*math.cos(math.radians(theta))

    ax = plt.gca()
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ax.set_xlim((-3,3))
    ax.set_ylim((-3,3))
    plt.gca().set_aspect('equal', adjustable='box')

    rect2 = patches.Rectangle((x1,y1), 2, 2, theta)
    ax.add_patch(rect2)
```
