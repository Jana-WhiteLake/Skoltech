```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
def new_value(sigma_x, sigma_y, tau_xy, theta, rou):
    plus = (sigma_x+sigma_y)/2
    minus = (sigma_x-sigma_y)/2
    sigma_x1 = round(plus + minus*math.cos(2*math.radians(theta))+tau_xy*math.sin(2*math.radians(theta)),rou)
    print("sigma_x1 =", sigma_x1)
    tau_x1y1 = round(-minus*math.sin(2*math.radians(theta))+tau_xy*math.cos(2*math.radians(theta)),rou)
    print("tau_x1y1 =", tau_x1y1)
    sigma_y1 = round(plus - minus*math.cos(2*math.radians(theta))-tau_xy*math.sin(2*math.radians(theta)),rou)
    print("sigma_y1 =", sigma_y1)
    return (plus, minus, sigma_x1, tau_x1y1, sigma_y1)
```
