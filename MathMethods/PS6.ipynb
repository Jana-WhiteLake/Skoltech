Let $x_0 = 2$: $$x_1 = x_0-\frac{f(x_0)}{f'(x_0)} = 2 - \frac{3}{4} = 1.25 $$

$$x_2 = x_1-\frac{f(x_1)}{f'(x_1)} = 1.25 - \frac{0.5625}{2.5} = 1.025 $$

$$x_3 = x_2-\frac{f(x_2)}{f'(x_2)} = 1.025 - \frac{0.050625}{2.05} = 1.0003 $$

$$x_4 = 1.00000005 $$

```python
def f(x):
    return(x**2 - 1)
n = 10
x_sh = np.empty(2*n+1, dtype=int)
x_sh[n] = 100000000 # division by 0 is not possible
x_n = 0
for i in range(-n, n+1):
    if (i==0): continue
    x_n = i
    x_sh[i+n] = x_n
    for j in range(n):
        x_np1 = x_n - f(x_n)/(2*x_n)
        x_n = x_np1
        if (np.abs(x_np1) < np.abs(x_sh[i+n])): 
            x_sh[i+n] = x_np1
        else:
            #kol[i+n] = j
            #e_n = np.abs(x_sh[i+n]-1)
            break
print(x_sh)
```

```python
x = np.arange(-10, 11, 1)
plt.ylim(-1.05,1.05)
plt.plot(x, x_sh, "ro", color='green')
plt.title('Convergence')
plt.xlabel('$x_0$')
plt.ylabel('$root$')
plt.grid()
```

For x greater than 0 converges to 1.
For x less than 0, converges to -1.
